```python
import pandas as pd
```

```python
df = pd.read_csv('bestsellers with categories.csv')
```

```python
print(df.head())
print(df.shape)
print(df.columns)
print(df.describe())
```

```python
df.drop_duplicates(inplace=True)
df.rename(columns={"Name":"Title","User Rating":"Public Rating","Year":"Publication year"})
```

```python
df['User Rating'] = df['User Rating'].astype(float)
print(df['User Rating'])
```

```python
author_book_counts = df['Author'].value_counts()
print(author_book_counts)
```

```python
Avg_rating_by_genre = df.groupby("Genre")["User Rating"].mean()
print(Avg_rating_by_genre)
```

```python
author_book_counts.head(10).to_csv("author_book_count.csv")
Avg_rating_by_genre.to_csv("Avg_rating_by_genre.csv")
```
